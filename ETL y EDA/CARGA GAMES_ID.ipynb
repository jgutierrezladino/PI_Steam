 ### Cargamos los datos del archivo games

```python
import pandas as pd
import gzip
import json
```

```python
steam_games = "../DataSets/steam_games.json.gz"
```

```python
datos_games = []
with gzip.open(steam_games, 'rt', encoding='utf-8') as archivo_games:
    for linea in archivo_games:
        diccionario_linea_games = json.loads(linea)
        datos_games.append(diccionario_linea_games)
df_games = pd.DataFrame(datos_games)
df_games.head()
```

```python
columnas_deseadas = ['id', 'genres', 'app_name', 'release_date']
df_games = df_games[columnas_deseadas]
```

```python
df_games = df_games.dropna()
```

```python
df_games_desa = df_games.explode('genres')
```

```python
df_desanidado = df_games_desa.reset_index(drop=True)
```

```python
df_desanidado.to_csv('../DataSets/df_games_id.csv',index=False)
```
